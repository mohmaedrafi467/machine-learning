```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Changed 'gaussiaan' to 'GaussianNB'
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix # Changed 'aaacuracy_score' to 'accuracy_score'
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
data = load_iris()
x = data.data
y = data.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
np_classifier = GaussianNB()
np_classifier.fit(x_train, y_train)
y_pred = np_classifier.predict(x_test)
conf_matrix = confusion_matrix(y_test, y_pred)
conf_report = classification_report(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred) # Defined 'accuracy' before using it
print("accuracy:", accuracy)
print("\n confusion matrix") # Changed '/n' to '\n' for a new line
print(conf_matrix)
print("\n classification report") # Changed '/n' to '\n' for a new line
print(conf_report)


# Removed redundant import statements

sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d", xticklabels=data.target_names, yticklabels=data.target_names)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title("Confusion Matrix Heatmap")
plt.show()
plt.title("confusion matrix heatmap") # Changed 'plf' to 'plt'
plt.show()
```
